```python
import pandas as pd
import time
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
df=pd.read_csv("zomato_table1.csv")
```

```python
df
```

```python
driver=webdriver.Chrome(executable_path="E:\\exe\\chromedriver.exe")
```

```python
delivery_reviews_num=[]
restaurant_address=[]
Latitude=[]
Longtitude=[]
Name=[]
```

```python
for i in range(550):
    time.sleep(1)
    driver.get(df.urls[i])
    time.sleep(1)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    review_tag = soup.find_all("div",class_="sc-1q7bklc-8 kEgyiI")
    reviewnum=review_tag[1].text
    
    address_tag=soup.find('a',class_="sc-clNaTc vNCcy")
    address=address_tag.text
    
    link_tag = soup.find('a', rel="noopener noreferrer")
    try:
        if 'href' in link_tag.attrs:
            link = link_tag.get('href')
    except:
        pass
    latitude=link[51:64]
    longtitude=link[65:]
    
    
    delivery_reviews_num.append(reviewnum)
    restaurant_address.append(address)
    Latitude.append(latitude)
    Longtitude.append(longtitude)
    Name.append(df.name[i])
```

```python
df2=pd.DataFrame({"Name":Name,"delivery_reviews_Count":delivery_reviews_num,"restaurant_address":restaurant_address,"Latitude":Latitude,"Longtitude":Longtitude})
```

```python
df2.to_csv("zomato_table2.csv")
```
